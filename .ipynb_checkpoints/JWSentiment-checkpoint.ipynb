```python
#SentimentSearchEngine

# Dependencies
import pandas as pd


from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#pass a CSV of tweets into df, run sentiment analysis
#Create a path to the csv and read it into a Pandas DataFrame


csv_path = "output/holdallthetweetsB.csv"

bey_df = pd.read_csv(csv_path, header=None)

bey_df.columns = ["Longitude","Latitude","Tweet Text", "City&State", "Lat&Long Together"]

# set up additional columns to hold information
#bey_df["Compound Sentiment"] = ""

bey_df.head()   

        
    
    
```

```python
#create a list to hold compound sentiments
compound_sentiments = []

# use iterrows to iterate through pandas dataframe
for index, row in bey_df.iterrows():
    
    target_sample = row["Tweet Text"]

    # Run Vader Analysis on each tweet
    results = analyzer.polarity_scores(target_sample)

    # Run analysis
    compound = results["compound"]
    #pos = results["pos"]
    #neu = results["neu"]
    #neg = results["neg"]
    
    # Add each value to the appropriate list
    compound_sentiments.append(compound)

    # Print Analysis
    #print(target_sample)
    #print("Compound Score:", compound)
    #print("Positive Score:", pos)
    #print("Neutral Score:", neu)
    #print("Negative Score: ", neg)
            
    # Print the Averages
    #print(f"User: {target_user}")
    #print(f"Compound: {np.mean(compound_sentiments):.3f}")
    
#print(compound_sentiments)        

            
```

```python
column_values = pd.Series(compound_sentiments)
bey_df.insert(loc=5, column="Compound Sentiment",value=column_values)

bey_df.head(15)
```
